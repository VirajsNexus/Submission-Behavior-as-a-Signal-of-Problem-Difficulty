## Step 1 — Install and Import Libraries

```python
# If any library is missing, run this cell first
# !pip install pandas numpy matplotlib seaborn scikit-learn xgboost
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import warnings
warnings.filterwarnings('ignore')

from sklearn.model_selection import train_test_split, cross_val_score
from sklearn.preprocessing   import StandardScaler
from sklearn.linear_model    import LogisticRegression
from sklearn.tree            import DecisionTreeClassifier
from sklearn.ensemble        import GradientBoostingClassifier
from sklearn.metrics         import (
    accuracy_score, f1_score, precision_score, recall_score,
    classification_report, confusion_matrix, ConfusionMatrixDisplay
)

# Plot settings
plt.rcParams.update({
    'figure.dpi'        : 120,
    'axes.spines.top'   : False,
    'axes.spines.right' : False,
    'font.family'       : 'DejaVu Sans'
})

RANDOM_STATE = 42
np.random.seed(RANDOM_STATE)

print('All libraries loaded successfully!')
```

---
## Step 2 — Load the Dataset

```python
# Load the CSV file
df = pd.read_csv('codeforces_dataset.csv')

print(f'Total Rows    : {df.shape[0]}')
print(f'Total Columns : {df.shape[1]}')
print()
print('Column Names:')
print(df.columns.tolist())
```

```python
# Preview first 5 rows
df.head()
```

```python
# Basic statistics for important columns
df[['acceptance_rate', 'wa_ratio', 'tle_ratio',
    'avg_solve_time_min', 'solver_rating_mean',
    'attempts_per_solver']].describe().round(3)
```

---
## Step 3 — Explore the Data (Charts)

```python
# ── Figure 1: Class Distribution ─────────────────────────
tier_counts = (
    df.groupby(['difficulty_tier', 'tier_name'])
      .size()
      .reset_index(name='count')
      .sort_values('difficulty_tier')
)

colors5 = ['#4CAF50', '#2196F3', '#FF9800', '#E91E63', '#9C27B0']

fig, ax = plt.subplots(figsize=(9, 5))
bars = ax.bar(
    tier_counts['tier_name'], tier_counts['count'],
    color=colors5, edgecolor='white', linewidth=1.5, width=0.6
)
for bar, count in zip(bars, tier_counts['count']):
    ax.text(
        bar.get_x() + bar.get_width() / 2,
        bar.get_height() + 10,
        str(count), ha='center', fontsize=11, fontweight='bold'
    )
ax.set_title('Figure 1 — Number of Problems per Difficulty Tier',
             fontsize=14, fontweight='bold', pad=12)
ax.set_xlabel('Difficulty Tier', fontsize=12)
ax.set_ylabel('Number of Problems', fontsize=12)
ax.set_ylim(0, tier_counts['count'].max() * 1.15)
plt.tight_layout()
plt.savefig('fig1_class_distribution.png', bbox_inches='tight')
plt.show()
print('Saved: fig1_class_distribution.png')
```

```python
# ── Figure 2: Acceptance Rate by Tier ────────────────────
tier_acc    = df.groupby('difficulty_tier')['acceptance_rate'].mean()
tier_labels = ['T1\nBeginner', 'T2\nEasy-Med',
               'T3\nMedium',   'T4\nHard', 'T5\nExpert']

fig, ax = plt.subplots(figsize=(9, 5))
bars = ax.bar(
    tier_labels, tier_acc.values,
    color=colors5, edgecolor='white', linewidth=1.5, width=0.55
)
for bar, val in zip(bars, tier_acc.values):
    ax.text(
        bar.get_x() + bar.get_width() / 2,
        bar.get_height() + 0.008,
        f'{val:.1%}', ha='center', fontsize=11, fontweight='bold'
    )
ax.set_title('Figure 2 — Mean Acceptance Rate by Difficulty Tier',
             fontsize=14, fontweight='bold', pad=12)
ax.set_xlabel('Difficulty Tier', fontsize=12)
ax.set_ylabel('Mean Acceptance Rate', fontsize=12)
ax.set_ylim(0, 0.90)
plt.tight_layout()
plt.savefig('fig2_acceptance_rate.png', bbox_inches='tight')
plt.show()
print('Saved: fig2_acceptance_rate.png')
```

```python
# ── Figure 3: Average Solving Time by Tier ────────────────
tier_time = df.groupby('difficulty_tier')['avg_solve_time_min'].mean()

fig, ax = plt.subplots(figsize=(9, 5))
ax.plot(
    tier_labels, tier_time.values,
    marker='o', color='#E91E63', linewidth=2.5,
    markersize=10, markerfacecolor='white', markeredgewidth=2.5
)
for i, (x, yv) in enumerate(zip(tier_labels, tier_time.values)):
    ax.annotate(
        f'{yv:.1f} min', (i, yv),
        textcoords='offset points', xytext=(0, 13),
        ha='center', fontsize=10
    )
ax.fill_between(range(5), tier_time.values, alpha=0.10, color='#E91E63')
ax.set_title('Figure 3 — Average Solving Time by Difficulty Tier',
             fontsize=14, fontweight='bold', pad=12)
ax.set_xticks(range(5))
ax.set_xticklabels(tier_labels)
ax.set_xlabel('Difficulty Tier', fontsize=12)
ax.set_ylabel('Avg Solve Time (minutes)', fontsize=12)
plt.tight_layout()
plt.savefig('fig3_solve_time.png', bbox_inches='tight')
plt.show()
print('Saved: fig3_solve_time.png')
```

---
## Step 4 — Preprocess the Data

```python
# Check for missing values
print('Missing values per column:')
missing = df.isnull().sum()
print(missing[missing > 0] if missing.any() else 'No missing values found!')
```

```python
# Drop non-feature columns
drop_cols = [
    'problem_id', 'contest_id', 'problem_index',
    'rating', 'difficulty_tier', 'tier_name', 'tags'
]
df_model = df.drop(columns=drop_cols, errors='ignore').copy()

# One-hot encode 'round_type'
df_model = pd.get_dummies(df_model, columns=['round_type'], drop_first=False)
print(f'Features after one-hot encoding: {df_model.shape[1]}')

# Log-transform skewed columns
for col in ['total_submissions', 'unique_solvers']:
    if col in df_model.columns:
        df_model[col] = np.log1p(df_model[col])
        print(f'  Log1p applied to: {col}')

# Define features (X) and target (y)
y            = df['difficulty_tier'].values   # labels: 1, 2, 3, 4, 5
X            = df_model.values.astype(float)
feature_names = df_model.columns.tolist()

print(f'\nFinal feature matrix shape: {X.shape}')
print(f'Target classes: {np.unique(y)}')
```

---
## Step 5 — Feature Engineering (New Features)

```python
# These 3 features were already created during data collection
# and are present in the CSV. Let's verify they exist.

engineered_features = ['solve_pressure', 'elite_selectivity', 'attempt_waste']

print('Engineered features check:')
for feat in engineered_features:
    if feat in df.columns:
        print(f'  ✓  {feat} — found in dataset')
        print(f'     mean = {df[feat].mean():.4f}, std = {df[feat].std():.4f}')
    else:
        print(f'  ✗  {feat} — NOT found, creating it now...')
        if feat == 'solve_pressure':
            df[feat] = df['avg_solve_time_min'] * (1 - df['acceptance_rate'])
        elif feat == 'elite_selectivity':
            df[feat] = (df['unique_solvers'] / df['total_submissions']) * (1 - df['acceptance_rate'])
        elif feat == 'attempt_waste':
            df[feat] = df['attempts_per_solver'] * df['wa_ratio']
        print(f'     Created successfully.')
```

---
## Step 6 — Train / Validation / Test Split

```python
# Split: 70% Train | 15% Validation | 15% Test
# Stratified = keeps class proportions same in all splits

X_temp, X_test,  y_temp, y_test  = train_test_split(
    X, y, test_size=0.15, stratify=y, random_state=RANDOM_STATE
)
X_train, X_val,  y_train, y_val  = train_test_split(
    X_temp, y_temp, test_size=0.1765, stratify=y_temp, random_state=RANDOM_STATE
)

print('Split Summary')
print(f'  Training   : {len(X_train):>5} problems  ({len(X_train)/len(X)*100:.1f}%)')
print(f'  Validation : {len(X_val):>5} problems  ({len(X_val)/len(X)*100:.1f}%)')
print(f'  Test       : {len(X_test):>5} problems  ({len(X_test)/len(X)*100:.1f}%)')

# Scale features using StandardScaler
# IMPORTANT: fit only on training data, then transform all splits
scaler    = StandardScaler()
X_train_s = scaler.fit_transform(X_train)   # fit + transform
X_val_s   = scaler.transform(X_val)         # transform only
X_test_s  = scaler.transform(X_test)        # transform only

print('\nStandardScaler applied. Features now have mean=0, std=1.')
```

---
## Step 7 — Baselines (Simple Rules Before ML)

```python
results = {}   # We'll store all model results here

# ── Baseline 1: Always predict the most common class ────
majority_class   = np.bincount(y_train).argmax()
y_pred_majority  = np.full(len(y_test), majority_class)

results['Majority Baseline'] = {
    'accuracy'    : accuracy_score(y_test, y_pred_majority),
    'macro_f1'    : f1_score(y_test, y_pred_majority, average='macro',  zero_division=0),
    'weighted_f1' : f1_score(y_test, y_pred_majority, average='weighted', zero_division=0),
}
print(f"Majority Baseline Accuracy : {results['Majority Baseline']['accuracy']:.1%}")

# ── Baseline 2: Simple acceptance rate threshold ─────────
acc_idx = feature_names.index('acceptance_rate')

def acceptance_rate_predict(acc_vals):
    preds = np.empty(len(acc_vals), dtype=int)
    for i, v in enumerate(acc_vals):
        if   v >= 0.55: preds[i] = 1
        elif v >= 0.35: preds[i] = 2
        elif v >= 0.20: preds[i] = 3
        elif v >= 0.10: preds[i] = 4
        else:           preds[i] = 5
    return preds

y_pred_thr = acceptance_rate_predict(X_test[:, acc_idx])

results['AccRate Baseline'] = {
    'accuracy'    : accuracy_score(y_test, y_pred_thr),
    'macro_f1'    : f1_score(y_test, y_pred_thr, average='macro',    zero_division=0),
    'weighted_f1' : f1_score(y_test, y_pred_thr, average='weighted', zero_division=0),
    'predictions' : y_pred_thr,
}
print(f"AccRate Baseline Accuracy  : {results['AccRate Baseline']['accuracy']:.1%}")
```

---
## Step 8 — Model 1: Logistic Regression

```python
print('Training Logistic Regression...')

lr_model = LogisticRegression(
    max_iter     = 2000,
    C            = 0.1,          # regularization strength
    multi_class  = 'ovr',        # one-vs-rest for multi-class
    random_state = RANDOM_STATE
)
lr_model.fit(X_train_s, y_train)

y_pred_lr = lr_model.predict(X_test_s)

results['Logistic Regression'] = {
    'accuracy'    : accuracy_score(y_test, y_pred_lr),
    'macro_f1'    : f1_score(y_test, y_pred_lr, average='macro'),
    'weighted_f1' : f1_score(y_test, y_pred_lr, average='weighted'),
    'predictions' : y_pred_lr,
}

print(f"Accuracy   : {results['Logistic Regression']['accuracy']:.1%}")
print(f"Macro-F1   : {results['Logistic Regression']['macro_f1']:.3f}")
print('Done!')
```

---
## Step 9 — Model 2: Decision Tree

```python
print('Training Decision Tree...')

dt_model = DecisionTreeClassifier(
    max_depth        = 7,    # limit depth to avoid overfitting
    min_samples_split = 20,  # need at least 20 samples to split a node
    min_samples_leaf  = 10,  # need at least 10 samples at each leaf
    random_state     = RANDOM_STATE
)
dt_model.fit(X_train_s, y_train)

y_pred_dt = dt_model.predict(X_test_s)

results['Decision Tree'] = {
    'accuracy'    : accuracy_score(y_test, y_pred_dt),
    'macro_f1'    : f1_score(y_test, y_pred_dt, average='macro'),
    'weighted_f1' : f1_score(y_test, y_pred_dt, average='weighted'),
    'predictions' : y_pred_dt,
}

print(f"Accuracy   : {results['Decision Tree']['accuracy']:.1%}")
print(f"Macro-F1   : {results['Decision Tree']['macro_f1']:.3f}")
print('Done!')
```

---
## Step 10 — Model 3: Gradient Boosting (Best Model)

```python
print('Training Gradient Boosting... (may take 1-2 minutes)')

gbm_model = GradientBoostingClassifier(
    n_estimators  = 300,   # number of trees
    max_depth     = 5,     # depth of each tree
    learning_rate = 0.1,   # how fast the model learns
    subsample     = 0.85,  # fraction of data used per tree
    random_state  = RANDOM_STATE
)
gbm_model.fit(X_train_s, y_train)

y_pred_gbm = gbm_model.predict(X_test_s)

results['Gradient Boosting'] = {
    'accuracy'    : accuracy_score(y_test, y_pred_gbm),
    'macro_f1'    : f1_score(y_test, y_pred_gbm, average='macro'),
    'weighted_f1' : f1_score(y_test, y_pred_gbm, average='weighted'),
    'predictions' : y_pred_gbm,
}

print(f"Accuracy   : {results['Gradient Boosting']['accuracy']:.1%}")
print(f"Macro-F1   : {results['Gradient Boosting']['macro_f1']:.3f}")
print('Done!')
```

---
## Step 11 — Results Summary Table

```python
# Build a clean summary DataFrame
summary_rows = []
for model_name, res in results.items():
    summary_rows.append({
        'Model'        : model_name,
        'Accuracy'     : f"{res['accuracy']:.1%}",
        'Macro-F1'     : f"{res['macro_f1']:.3f}",
        'Weighted-F1'  : f"{res['weighted_f1']:.3f}",
    })

summary_df = pd.DataFrame(summary_rows)
summary_df.index = summary_df.index + 1  # start index from 1
summary_df
```

---
## Step 12 — Figure 4: Model Accuracy Comparison

```python
model_names  = list(results.keys())
accuracies   = [results[m]['accuracy'] * 100 for m in model_names]
bar_colors   = ['#B0BEC5', '#90A4AE', '#42A5F5', '#66BB6A', '#EF5350']

fig, ax = plt.subplots(figsize=(10, 5))
bars = ax.barh(
    model_names, accuracies,
    color=bar_colors, edgecolor='white', linewidth=1.5, height=0.55
)
for bar, val in zip(bars, accuracies):
    ax.text(
        val + 0.5, bar.get_y() + bar.get_height() / 2,
        f'{val:.1f}%', va='center', fontsize=11, fontweight='bold'
    )
ax.set_xlim(0, 100)
ax.set_xlabel('Accuracy (%)', fontsize=12)
ax.set_title('Figure 4 — Model Accuracy Comparison',
             fontsize=14, fontweight='bold', pad=12)
ax.axvline(x=50, color='gray', linewidth=1, linestyle='--', alpha=0.5)
plt.tight_layout()
plt.savefig('fig4_model_accuracy.png', bbox_inches='tight')
plt.show()
print('Saved: fig4_model_accuracy.png')
```

---
## Step 13 — Figure 5: Confusion Matrix

```python
tier_display = ['Beginner', 'Easy-Med', 'Medium', 'Hard', 'Expert']

cm = confusion_matrix(y_test, y_pred_gbm)

fig, ax = plt.subplots(figsize=(8, 6))
sns.heatmap(
    cm, annot=True, fmt='d', cmap='Blues',
    xticklabels=tier_display, yticklabels=tier_display,
    linewidths=0.5, ax=ax, annot_kws={'size': 12}
)
ax.set_title('Figure 5 — Confusion Matrix (Gradient Boosting)',
             fontsize=14, fontweight='bold', pad=12)
ax.set_xlabel('Predicted Label', fontsize=12)
ax.set_ylabel('True Label', fontsize=12)
plt.tight_layout()
plt.savefig('fig5_confusion_matrix.png', bbox_inches='tight')
plt.show()
print('Saved: fig5_confusion_matrix.png')
print()
print('Note: Diagonal = correct predictions.')
print('      Most errors are between neighbouring tiers (very good!).')
```

---
## Step 14 — Detailed Classification Report

```python
print('=' * 55)
print('  Classification Report — Gradient Boosting Model')
print('=' * 55)
print(classification_report(
    y_test, y_pred_gbm,
    target_names=tier_display
))
```

---
## Step 15 — Figure 6: Feature Importance

```python
# Extract feature importances from the GBM model
fi_df = pd.DataFrame({
    'Feature'    : feature_names,
    'Importance' : gbm_model.feature_importances_
})
fi_df = fi_df.sort_values('Importance', ascending=False).head(12)

print('Top 12 Most Important Features:')
print(fi_df.to_string(index=False))
```

```python
# Plot feature importances
fi_sorted = fi_df.sort_values('Importance')   # ascending for horizontal bar
n         = len(fi_sorted)
fi_colors = plt.cm.RdYlGn(np.linspace(0.3, 0.9, n))

fig, ax = plt.subplots(figsize=(10, 6))
ax.barh(
    fi_sorted['Feature'], fi_sorted['Importance'],
    color=fi_colors, edgecolor='white', linewidth=1.2, height=0.7
)
ax.set_xlabel('Feature Importance Score', fontsize=12)
ax.set_title('Figure 6 — Top Feature Importances (Gradient Boosting)',
             fontsize=14, fontweight='bold', pad=12)
for spine in ax.spines.values():
    spine.set_visible(False)
plt.tight_layout()
plt.savefig('fig6_feature_importance.png', bbox_inches='tight')
plt.show()
print('Saved: fig6_feature_importance.png')
```

---
## Step 16 — Figure 7: Per-Class F1 for All Models

```python
ml_models   = ['Logistic Regression', 'Decision Tree', 'Gradient Boosting']
per_class_f1 = {}

for mk in ml_models:
    preds = results[mk]['predictions']
    per_class_f1[mk] = f1_score(
        y_test, preds, average=None, labels=[1, 2, 3, 4, 5]
    )

# Build a readable table
pc_df = pd.DataFrame(per_class_f1, index=tier_display).round(3)
print('Per-Class F1 Score Comparison:')
pc_df
```

```python
x      = np.arange(5)
width  = 0.25
m_colors = ['#42A5F5', '#66BB6A', '#EF5350']

fig, ax = plt.subplots(figsize=(11, 5))
for i, (mk, clr) in enumerate(zip(ml_models, m_colors)):
    ax.bar(
        x + i * width, per_class_f1[mk], width,
        label=mk, color=clr, edgecolor='white', linewidth=1.2
    )
ax.set_xticks(x + width)
ax.set_xticklabels(tier_display, fontsize=11)
ax.set_ylabel('F1 Score', fontsize=12)
ax.set_ylim(0, 1.12)
ax.set_title('Figure 7 — Per-Class F1 Score by Model',
             fontsize=14, fontweight='bold', pad=12)
ax.legend(fontsize=10)
plt.tight_layout()
plt.savefig('fig7_perclass_f1.png', bbox_inches='tight')
plt.show()
print('Saved: fig7_perclass_f1.png')
```

---
## Step 17 — Figure 8: Feature Correlation Heatmap

```python
num_cols = [
    'acceptance_rate', 'wa_ratio', 'tle_ratio',
    'avg_solve_time_min', 'median_solve_time_min',
    'solver_rating_mean', 'attempts_per_solver',
    'solve_pressure', 'attempt_waste'
]

# Use only columns that actually exist in df
existing_cols = [c for c in num_cols if c in df.columns]
corr = df[existing_cols + ['difficulty_tier']].corr()

fig, ax = plt.subplots(figsize=(11, 8))
sns.heatmap(
    corr, annot=True, fmt='.2f', cmap='coolwarm',
    center=0, linewidths=0.5, ax=ax,
    annot_kws={'size': 8}
)
ax.set_title('Figure 8 — Feature Correlation Heatmap',
             fontsize=14, fontweight='bold', pad=12)
plt.tight_layout()
plt.savefig('fig8_correlation_heatmap.png', bbox_inches='tight')
plt.show()
print('Saved: fig8_correlation_heatmap.png')
```

---
## Step 18 — Cross Validation (Extra Reliability Check)

```python
print('Running 5-Fold Cross Validation on Gradient Boosting...')
print('(This may take 2-3 minutes)')

cv_model = GradientBoostingClassifier(
    n_estimators=300, max_depth=5,
    learning_rate=0.1, subsample=0.85,
    random_state=RANDOM_STATE
)
cv_scores = cross_val_score(
    cv_model, X_train_s, y_train,
    cv=5, scoring='accuracy', n_jobs=-1
)

print(f'\nCV Scores per fold : {[f"{s:.3f}" for s in cv_scores]}')
print(f'Mean CV Accuracy   : {cv_scores.mean():.3f}')
print(f'Std Deviation      : {cv_scores.std():.3f}')
print()
print('Low std deviation = the model is consistent, not just lucky!')
```

```python
print('=' * 60)
print('  FINAL RESULTS SUMMARY')
print('=' * 60)

for m, r in results.items():
    print(f"  {m:<25}  Accuracy={r['accuracy']:.1%}   "
          f"MacroF1={r['macro_f1']:.3f}   "
          f"WeightedF1={r['weighted_f1']:.3f}")

best_acc  = results['Gradient Boosting']['accuracy']
best_f1   = results['Gradient Boosting']['macro_f1']
top_feat  = fi_df.iloc[0]['Feature']

print()
print(f'  Best Model         : Gradient Boosting')
print(f'  Best Accuracy      : {best_acc:.1%}')
print(f'  Best Macro-F1      : {best_f1:.3f}')
print(f'  Top Feature        : {top_feat}')
print()
print('  Figures Saved:')
for i in range(1, 9):
    names = [
        'Class Distribution', 'Acceptance Rate', 'Solve Time',
        'Model Accuracy', 'Confusion Matrix', 'Feature Importance',
        'Per-Class F1', 'Correlation Heatmap'
    ]
    print(f'    fig{i}_{names[i-1].lower().replace(" ","_")}.png')
print('=' * 60)
```
